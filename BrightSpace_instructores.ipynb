```python
# Paso 2: Importar librer칤as
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import gspread
from gspread_dataframe import get_as_dataframe
from google.colab import auth
from google.auth import default
```

```python
# Paso 3: Autenticaci칩n con Google
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
# Paso 4: Leer Google Sheet
SHEET_ID = "1Eh3X8Bwd-0GoF3pyl0Xz6-oCOJm-SW9l5fItahcn99g"
worksheet = gc.open_by_key(SHEET_ID).sheet1
df = get_as_dataframe(worksheet, evaluate_formulas=True).dropna(how="all")
```

```python
# Paso 5: Limpieza y conversi칩n
cols_metrica = [
    "Cantidad de elementos de calificaci칩n",
    "Cantidad de publicaciones de debate",
    "Cantidad de publicaciones de debate iniciado",
    "Cantidad de inicios de sesi칩n en el sistema"
]
df[cols_metrica] = df[cols_metrica].apply(pd.to_numeric, errors='coerce').fillna(0)
```

```python
# Paso 6: Tabla resumen
resumen_total = df[cols_metrica].sum().reset_index()
resumen_total.columns = ["M칠trica", "Total"]
fig_tabla = go.Figure(data=[go.Table(
    header=dict(values=["M칠trica", "Total"], fill_color='lightblue', align='left'),
    cells=dict(values=[resumen_total["M칠trica"], resumen_total["Total"]], fill_color='lavender', align='left')
)])
fig_tabla.update_layout(title="Resumen General de Actividades desde Google Sheets")
fig_tabla.show()
```

```python
# Paso 7: Indicadores KPI
fig_kpi = go.Figure()
fig_kpi.add_trace(go.Indicator(mode="number", value=resumen_total.loc[0, "Total"], title="游닇 Calificaciones", domain={'x': [0, 0.25], 'y': [0, 1]}))
fig_kpi.add_trace(go.Indicator(mode="number", value=resumen_total.loc[1, "Total"], title="游눫 Publicaciones de debate", domain={'x': [0.26, 0.5], 'y': [0, 1]}))
fig_kpi.add_trace(go.Indicator(mode="number", value=resumen_total.loc[2, "Total"], title="游닉 Debates iniciados", domain={'x': [0.51, 0.75], 'y': [0, 1]}))
fig_kpi.add_trace(go.Indicator(mode="number", value=resumen_total.loc[3, "Total"], title="游댏 Inicios de sesi칩n", domain={'x': [0.76, 1], 'y': [0, 1]}))
fig_kpi.update_layout(title="Indicadores Clave desde Google Sheets")
fig_kpi.show()
```

```python
# Paso 8: Gr치ficos por profesor (con etiquetas visibles)
fig1 = px.bar(df, x="Nombre de Profesor", y="Cantidad de elementos de calificaci칩n",
              title="游닇 Elementos de calificaci칩n por profesor", color="Nombre de Profesor", text_auto=True)
fig1.show()
fig2 = px.bar(df, x="Nombre de Profesor", y="Cantidad de publicaciones de debate",
              title="游눫 Publicaciones de debate por profesor", color="Nombre de Profesor", text_auto=True)
fig2.show()
fig3 = px.bar(df, x="Nombre de Profesor", y="Cantidad de publicaciones de debate iniciado",
              title="游닉 Debates iniciados por profesor", color="Nombre de Profesor", text_auto=True)
fig3.show()
fig4 = px.bar(df, x="Nombre de Profesor", y="Cantidad de inicios de sesi칩n en el sistema",
              title="游댏 Inicios de sesi칩n por profesor", color="Nombre de Profesor", text_auto=True)
fig4.show()
```
